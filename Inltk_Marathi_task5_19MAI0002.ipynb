Task 5 : Applications for the functionalities available in iNLTK for regional language (Marathi here)

Language setup

```python
from inltk.inltk import setup
```

```python
setup('mr')
```

Importing and reading 'Marathi' text

```python
file = open('/content/marathi.txt','r').read()
```

```python
file
```

Tokenization

```python
from inltk.inltk import tokenize
```

```python
print(tokenize(file,'mr'),end=" ")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

Identification of language

```python
from inltk.inltk import identify_language
```

```python
identify_language("मराठा साम्राज्याचे संस्थापक")
```

Embedding vectors

iNLTK under the hood utilizes the ULMFiT method of training language models and hence it can generate vector embeddings for a given input text

```python
from inltk.inltk import get_embedding_vectors
```

```python
embedd_vectors = get_embedding_vectors(file,'mr')
```

```python
embedd_vectors
```

```python
print("Number of vectors:", len(embedd_vectors))
print("Shape of each vector:", len(embedd_vectors[0]))
```

Completing text from predicting next words

```python
from inltk.inltk import predict_next_words
```

```python
sentence = "छत्रपती शिवाजी प्रतिहल्ला करणाऱ्या" 
#छत्रपती शिवाजी प्रतिहल्ला करणाऱ्या हिंदू राष्ट्रवादाचे प्रतीक होते- original sentence
```

```python
predict_next_words(sentence,5,'mr')
```

```python
predict_next_words(sentence,10,'mr')
```

Get similar sentences from input

```python
from inltk.inltk import get_similar_sentences
```

```python
get_similar_sentences(sentence,5,'mr')
```

```python
get_similar_sentences('मराठ्यांचा राज्यविस्तार होतच राहिला', 5, 'mr')
```

Finding cosine similarities from given sentences

```python
from inltk.inltk import get_sentence_similarity
get_sentence_similarity('मराठ्यांचा राज्यविस्तार होतच राहिला', 'किनारी आणि अंतर्गत प्रदेशातील किल्ल्यांची डागडुजी करण्यासोबतच अनेक किल्लेही त्यांनी उभारले', 'mr')
```
