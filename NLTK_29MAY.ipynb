```python
# task 3 stemming paragraphs

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
example = 'The Natural Language Toolkit, or more commonly NLTK, is a suite of libraries and programs for symbolic and statistical natural language processing for English written in the Python programming language'
example = [stemmer.stem(token) for token in example.split(" ")]
print(" ".join(example))
```

```python
# task 6: basic text processing pipeline
from nltk.tokenize import sent_tokenize, word_tokenize
txt = "Hello Mr. Smith, how are you doing today? The weather is great, and Python is awesome. The sky is pinkish-blue. You shouldn't eat cardboard."
print(word_tokenize(txt))
```

```python
print(sent_tokenize(txt))
```

```python
for t in txt:
    sentences = nltk.sent_tokenize(txt)
    for s in sentences:
        words = nltk.word_tokenize(s)
        print(words)
        tags = nltk.pos_tag(words)
        print(tags)
```
