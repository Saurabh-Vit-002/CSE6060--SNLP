### Training with already tagged words to tag custom para

```python
# 5)
from nltk.tag import tnt
import string
```

```python
tagged_set = 'marathi.pos'
word_set = indian.sents(tagged_set)
count = 0
for sen in word_set[:100]:
    count = count + 1
    sen = "".join([" "+i if not i.startswith("'") and i not in string.punctuation else i for i in sen]).strip()
    print (count, sen)
```

```python
print ('Total no of sentences in the tagged file are: ',count)
train_perc = .9

train_rows = int(train_perc*count)
test_rows = train_rows + 1

print ('No of sentences to be trained: ',train_rows,'& No of sentences to be tested against: ',test_rows)
```

```python
data = indian.tagged_sents(tagged_set)
train_data = data[:train_rows] 
test_data = data[test_rows:]


pos_tagger = tnt.TnT()
pos_tagger.train(train_data)
pos_tagger.evaluate(test_data)
```

### Tokenization

```python
# 6) Tokenization 
tokenize1 = nltk.word_tokenize(para)
tokenize1[:100]
```

```python
tokenize2 = nltk.sent_tokenize(para)
tokenize2[:100]
```

### POS-Tagging

```python
# 7) pos tagging
for sent in tokenize2:
    tokenized_text = nltk.word_tokenize(sent)
    tagged = nltk.pos_tag(tokenized_text)
    print(tagged,"\n")
```

### StopWords Removal in Marathi

```python
# 8) Removing stopwords
stop_words = open('C:/Users/HP/Downloads/marathi_stopwords.txt','r',encoding="utf-8").read()
```

```python
print(stop_words)
```

```python
len(tokenize1)  # before removing stop words
```

```python
# Removing stop words

new_string = list()
for word in tokenize1:
    if word not in stop_words:
        new_string.append(word)
```

### Bigram

```python
# 9) Displaying bigrams
bigram = list(nltk.bigrams(tokenize1))
bigram[:100]
```

### Collections

```python
# 10) Extracting collections
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures 
biagram_collocation = BigramCollocationFinder.from_words(bigram)
```

```python
biagram_collocation.nbest(BigramAssocMeasures.likelihood_ratio, 30)
```
