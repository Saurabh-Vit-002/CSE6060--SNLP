```python
from nltk.corpus import stopwords
```

```python
count = 0
for i in range(len(sentences)):
    word_token = nltk.word_tokenize(sentences[i])
    words = [word for word in word_token if word not in stopwords.words('english')]   #task 2
    sentences[i] = ' '.join(words)
    count += len(words)     #no of words after removing stopwords
count
```

```python
# task 3
tokens = nltk.word_tokenize(txt)
bigram = list(nltk.bigrams(tokens)) 
```

```python
bigram
```

```python
# task 4
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures 
```

```python
biagram_collocation = BigramCollocationFinder.from_words(bigram)
```

```python
biagram_collocation.nbest(BigramAssocMeasures.likelihood_ratio, 30)
```

```python
import pandas as pd
bigram_freq = biagram_collocation.ngram_fd.items()
bigramFreqTable = pd.DataFrame(list(bigram_freq), columns=['bigram','freq']).sort_values(by='freq', ascending=False)
bigramFreqTable
```
